```python
data = pd.read_csv("/content/DS3001_project/StudentPerformanceFactors.csv") # use pd to refer to pandas
data.head() # show the first 5 rows of the data
```

```python
print(data.columns)
```

```python
plt.hist(data['Attendance'], bins=10)
plt.xlabel('Attendance')
plt.ylabel('Frequency')
plt.title('Histogram of Attendance')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(data['Sleep_Hours'],bins=6)  # Adjust bins as needed
plt.xlabel('Sleep Hours')
plt.ylabel('Frequency')
plt.title('Histogram of Sleep Hours')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x='Parental_Involvement', y='Exam_Score', data=data)
plt.xlabel('Parental Involvement')
plt.ylabel('Test Score')
plt.title('Box Plot of Test Score vs. Parental Involvement')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Learning_Disabilities', y='Exam_Score', data=data)
plt.xlabel('Learning_Disabilities')
plt.ylabel('Test Score')
plt.title('Box Plot of Test Score vs. Learning_Disabilities')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Family_Income', y='Exam_Score', data=data)
plt.xlabel('Family_Income')
plt.ylabel('Test Score')
plt.title('Box Plot of Test Score vs. Family_Income')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sleep_Hours', y='Exam_Score', data=data)
plt.xlabel('Sleep_Hours')
plt.ylabel('Test Score')
plt.title('Box Plot of Test Score vs. Sleep_Hours')
plt.show()
```
