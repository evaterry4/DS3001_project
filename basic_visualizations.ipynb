```python
import pandas as pd
data = pd.read_csv("/content/DS3001_project/college_data_2022_23 (1).csv")
```

```python
#I used the built in ai feature to quickly convert all the nummeric variables to numeric
numeric_columns = data.select_dtypes(include=['number']).columns

for column in numeric_columns:
  data[column] = pd.to_numeric(data[column], errors='coerce')

print(data.dtypes)
```

```python
data['FAMINC'] = pd.to_numeric(data['FAMINC'], errors='coerce')
print(data['FAMINC'].dtype)
```

```python
data['FEMALE'] = pd.to_numeric(data['FEMALE'], errors='coerce')
print(data['FEMALE'].dtype)
```

```python
# correlation heatmap for the numeric variables

# get only numeric variables
numeric_data = data.select_dtypes(include='number')

# make correlation heatmap
# https://www.geeksforgeeks.org/how-to-create-a-seaborn-correlation-heatmap-in-python/


# variable names to their descriptions
var_descriptions = {
    'ADM_RATE': 'Acceptance Rate',
    'PPTUG_EF': 'Part-Time Undergrad Share',
    'GRAD_DEBT_MDN_SUPP': 'Median Total Debt After Graduation',
    'PREDDEG': 'Most Common Degree Offered',
    'PCTPELL_DCS_POOLED_SUPP': 'Percentage of Pell Grant Recipients',
    'STUFACR': 'Student/Faculty Ratio',
    'PCTFLOAN_DCS_POOLED_SUPP': 'Proportion Borrowed Federal Loans',
    'UNITID': 'School ID',
    'INSTNM': 'School Name',
    'ZIP': 'ZIP Code',
    'STABBR': 'State Abbreviation',
    'C150_4': 'Completion Rate (within 150%)',
    'COSTT4_A': 'Yearly Cost of Attendance',
    'UGDS': 'Undergraduate Enrollment',
    'MD_EARN_WNE_4YR': 'Median Earnings (4 Years After Grad)',
    'CONTROL_PEPS': 'Control: Public vs. Private',
    'REGION': 'Region',
    'ACTCMMID': 'Midpoint ACT Score',
    'FEMALE': 'Percentage of Female Students',
    'FAMINC': 'Average Family Income'
}

correlation_matrix = numeric_data.corr()
correlation_matrix.rename(index=var_descriptions, columns=var_descriptions, inplace=True)
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='viridis', center=0)
plt.title('Correlation Heatmap of Variables')
plt.xlabel("Variable")
plt.ylabel("Variable")
plt.show()
```
