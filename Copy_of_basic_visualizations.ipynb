```python
import pandas as pd
data = pd.read_csv("/content/college_data_2022_23 (1).csv")
```

```python
data
```

```python
#I used the built in ai feature to quickly convert all the nummeric variables to numeric
numeric_columns = data.select_dtypes(include=['number']).columns

for column in numeric_columns:
  data[column] = pd.to_numeric(data[column], errors='coerce')

print(data.dtypes)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Histogram of Cost of attendance
sns.histplot(data['COSTT4_A'], bins=20)
plt.title('Distribution of Average Cost of Attendance')
plt.xlabel('Average Cost of Attendance')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram of Student Faculty Ratio
plt.figure(figsize=(8, 6))
sns.histplot(data['STUFACR'], bins=20)
plt.title('Distribution of Student Faculty Ratio')
plt.xlabel('Student Faculty Ratio')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of SAT score vs. Graduation rate
sns.scatterplot(x='ACTCMMID', y='C150_4', data=data)
plt.title('ACT Average vs. Graduation Rate')
plt.xlabel('Average ACT Score')
plt.ylabel('Graduation Rate')
plt.show()

# Scatter plot of Cost of attendance vs. Graduation rate
sns.scatterplot(x='COSTT4_A', y='C150_4', data=data)
plt.title('Cost of Attendance vs. Graduation Rate')
plt.xlabel('Average Cost of Attendance')
plt.ylabel('Graduation Rate')
plt.show()

# Scatter plot of acceptance rate vs. Graduation rate
sns.scatterplot(x='ADM_RATE', y='C150_4', data=data)
plt.title('Acceptance Rate vs. Graduation Rate')
plt.xlabel('Acceptance Rate')
plt.ylabel('Graduation Rate')
plt.show()
```

```python
# Scatter plot of admission rate vs. enrollment
sns.scatterplot(x='ADM_RATE', y='UGDS', data=data)
plt.title('Acceptance Rate vs. Undergraduate Enrollment')
plt.xlabel('Admission Rate')
plt.ylabel('Undergraduate Enrollment')
plt.show()
```

```python
data['PREDDEG_Description'] = data['PREDDEG'].map({
    1: 'Predominantly certificate-degree',
    2: 'Predominantly associate\'s-degree',
    3: 'Predominantly bachelor\'s-degree',
    4: 'Entirely graduate-degree'
})
```

```python
sns.boxplot(x='PREDDEG_Description', y='COSTT4_A', data=data)
plt.title('Cost of Attendance by Predominant Degree')
plt.xlabel('Predominant Degree Awarded')
plt.xticks(rotation=90)
plt.ylabel('Average Cost of Attendance')
plt.show()
```

```python
sns.boxplot(x='PREDDEG_Description', y='MD_EARN_WNE_4YR', data=data)
plt.title('Average Earnings by Predominant Degree')
plt.xlabel('Predominant Degree Awarded')
plt.xticks(rotation=90)
plt.ylabel('Average earnings 4 years after graduation')
plt.show()
```

```python
# I used the built in ai to Create a dictionary to map region codes to region names using the data dictionary to know the levels
region_mapping = {
    0: 'U.S. Service Schools',
    1: 'New England',
    2: 'Mid East',
    3: 'Great Lakes',
    4: 'Plains',
    5: 'Southeast',
    6: 'Southwest',
    7: 'Rocky Mountains',
    8: 'Far West'
}

# Map the region codes to region names in the DataFrame
data['REGION_Description'] = data['REGION'].map(region_mapping)

# Example: Box plot of cost of attendance by region
sns.boxplot(x='REGION_Description', y='COSTT4_A', data=data)
plt.title('Cost of Attendance by Region')
plt.xlabel('Region')
plt.xticks(rotation=90)
plt.ylabel('Average Cost of Attendance')
plt.show()
```

```python
#Scatter Plot Student Teacher Ratios and Graduation Rates

import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(x='STUFACR', y='C150_4', data=data)
plt.title('Student Faculty Ratio vs. Graduation Rate')
plt.xlabel('Student Faculty Ratio')
plt.ylabel('Graduation Rate')
plt.show()
```

```python
region_mapping = {
    0: 'U.S. Service Schools',
    1: 'New England',
    2: 'Mid East',
    3: 'Great Lakes',
    4: 'Plains',
    5: 'Southeast',
    6: 'Southwest',
    7: 'Rocky Mountains',
    8: 'Far West'
}

# Map the region codes to region names in the DataFrame
data['REGION_Description'] = data['REGION'].map(region_mapping)

sns.boxplot(x='REGION_Description', y='STUFACR', data=data)
plt.title('Student Faculty Ratio by Region')
plt.xlabel('Region')
plt.xticks(rotation=90)
plt.ylabel('Student Faculty Ratio')
plt.show()
#
```

```python
#cost of attendance by student faculty ratio and mixed with graduation rates
var1 = 'STUFACR'
var2 = 'COSTT4_A'
sns.jointplot(x=var1, y=var2, data=data, kind='scatter', color='blue')
plt.xlabel('Student Faculty Ratio')
plt.ylabel('Cost for Attendance')
plt.show()
```
