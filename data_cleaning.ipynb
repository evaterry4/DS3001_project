```python
import pandas as pd
data = pd.read_csv("/content/DS3001_project/college_data_2022_23 (1).csv")
```

```python
import matplotlib.pyplot as plt
plt.hist(data['PREDDEG'], bins=10)  # Adjust bins as needed
plt.xlabel('preddeg')
plt.ylabel('Frequency')
plt.title('Histogram of preddeg')
plt.show()
```

```python
#filter to only include schools that primarily offer 4-year bachelor's degrees
data2 = data[data['PREDDEG'] == 3]
data2
```

```python
numeric_columns = data2.select_dtypes(include=['number']).columns

for column in numeric_columns:
  data2[column] = pd.to_numeric(data2[column], errors='coerce')

print(data2.dtypes)
```

```python
print(data2['C150_4'].isnull().sum())
print(len(data2))
```

```python
data3 = data2.rename(columns={'UNITID':'id', 'INSTNM':'name', 'ZIP':'zip','CITY':'city',
'STABBR':'state', 'ADM_RATE':'acceptance_rate', 'PPTUG_EF':'part_time','GRAD_DEBT_MDN_SUPP':'debt',
'PREDDEG':'degree_type', 'PCTPELL_DCS_POOLED_SUPP':'pell_grant', 'STUFACR':'stud_fac_ratio','PCTFLOAN_DCS_POOLED_SUPP':'percent_loans',
'C150_4':'graduation_rate', 'COSTT4_A':'yearly_cost', 'UGDS':'student_pop','MD_EARN_WNE_4YR':'earnings',
'CONTROL_PEPS':'school_type', 'REGION':'region', 'ACTCMMID':'average_act','FEMALE':'precent_female','FAMINC':'average_income'
})
```
