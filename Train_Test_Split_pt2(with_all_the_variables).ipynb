```python
#clean the reigon variable based on the data dictionary
region_mapping = {
    0: 'US-service-schools',
    1: 'new-england',
    2: 'mid-east',
    3: 'great-lakes',
    4: 'plains',
    5: 'southeast',
    6: 'southwest',
    7: 'rocky-mountains',
    8: 'far-west'
}

# Map the region codes to region names in the DataFrame
data['region'] = data['region'].map(region_mapping)
```

```python
data_clean = data.dropna()
y = data_clean['graduation_rate']

# I belive we have to one-hot encode region before the train test split
df = pd.get_dummies(data_clean[['region', 'acceptance_rate','average_act', 'pell_grant','stud_fac_ratio','percent_loans','average_act', 'average_income']],
                   columns=['region'],
                   drop_first=False)  # Keep all categories for interpretability

# Split the sample:
from sklearn.model_selection import train_test_split
df_train, df_test, y_train, y_test = train_test_split(df, y,
                                                    test_size=.2,
                                                    random_state=100)
```

```python
from sklearn import linear_model
#ols model with all the variables


#X_train1 = pd.get_dummies(df_train, columns=['region', 'acceptance_rate','average_act', 'pell_grant','stud_fac_ratio','percent_loans','average_act', 'average_income'], drop_first=False)
#X_test1 = pd.get_dummies(df_test, columns=['region', 'acceptance_rate','average_act', 'pell_grant','stud_fac_ratio','percent_loans','average_act', 'average_income'], drop_first=False)

reg = linear_model.LinearRegression(fit_intercept=False).fit(df_train,y_train)


y_hat = reg.predict(df_test)

r2_score = reg.score(df_test, y_test)
print(f"R squared score: {r2_score}")

rmse = np.sqrt( np.mean( (y_test - y_hat)**2 ))
print(f"RMSE: {rmse}")

results = pd.DataFrame({'Variable': reg.feature_names_in_, 'Coefficient': reg.coef_})
print(results)
```
